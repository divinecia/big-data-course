```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Create a sample DataFrame (replace with your actual data loading)
data = {'Size': [14, 15, 17, 19, 16], 'Price': [178, 230, 240, 275, 200]}
df = pd.DataFrame(data)

# Print column names and head for verification
print(df.columns)
print(df.head())

# Data Cleaning (Important!)
# Check for missing values
print("\nMissing Values:")
print(df.isnull().sum())  # Shows the number of missing values per column
# df.dropna(inplace=True)  # Drop rows with any missing values (if needed)
# df['Size'].fillna(df['Size'].mean(), inplace=True) # Fill missing 'Size' with mean (if needed)
# df['Price'].fillna(df['Price'].mean(), inplace=True) # Fill missing 'Price' with mean (if needed)

# Check for duplicates
print("\nDuplicate Rows:")
print(df.duplicated().sum()) # Shows the number of duplicate rows
# df.drop_duplicates(inplace=True) # Drop duplicate rows (if needed)


# Calculate statistics
price_mean = df['Price'].mean()
price_std = df['Price'].std()
price_min = df['Price'].min()
price_max = df['Price'].max()
price_q25 = df['Price'].quantile(0.25)
price_q50 = df['Price'].quantile(0.50)
price_q75 = df['Price'].quantile(0.75)

size_mean = df['Size'].mean()
size_std = df['Size'].std()
size_min = df['Size'].min()
size_max = df['Size'].max()
size_q25 = df['Size'].quantile(0.25)
size_q50 = df['Size'].quantile(0.50)
size_q75 = df['Size'].quantile(0.75)

print("\nPrice Statistics:")
print(f"Mean: {price_mean}")
print(f"Standard Deviation: {price_std}")
print(f"Min: {price_min}")
print(f"Max: {price_max}")
print(f"25th Quantile: {price_q25}")
print(f"Median (50th Quantile): {price_q50}")
print(f"75th Quantile: {price_q75}")

print("\nSize Statistics:")
print(f"Mean: {size_mean}")
print(f"Standard Deviation: {size_std}")
print(f"Min: {size_min}")
print(f"Max: {size_max}")
print(f"25th Quantile: {size_q25}")
print(f"Median (50th Quantile): {size_q50}")
print(f"75th Quantile: {size_q75}")



# Calculate correlation coefficient
correlation = df['Price'].corr(df['Size'])
print(f"\nCorrelation coefficient between Price and Size: {correlation}")

# Visualize data
plt.figure(figsize=(12, 6))

# Scatter plot
plt.subplot(1, 2, 1)
plt.scatter(df['Size'], df['Price'])
plt.title('Scatter Plot of Price vs. Size')
plt.xlabel('Size')
plt.ylabel('Price')

# Box plot of Price
plt.subplot(1, 2, 2)
sns.boxplot(y=df['Price'])
plt.title('Box Plot of Price')
plt.ylabel('Price')

plt.tight_layout()
plt.show()

# Box plot of Size
plt.figure(figsize=(6, 6))
sns.boxplot(y=df['Size'])
plt.title('Box Plot of Size')
plt.ylabel('Size')
plt.show()

# Correlation Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
